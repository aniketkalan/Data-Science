Flask -  It is a lightweight web framework designed to create web applications

Q3. What is App routing in Flask? Why do we use app routes?
Ans : 
App Routing : It refers to the process of defining URL routes and associating them with specific functions or view handlers within a Flask application. It determines how different URL paths are mapped to corresponding functions that handle the incoming requests and generate the appropriate responses.

Why do we use app routes:
1. URL Mapping: App routes allow us to map specific URLs or URL patterns to corresponding functions or view handlers within the application. 

2. Request Handling: By using app routes, we can specify different functions to handle various types of requests such as GET, POST, PUT, DELETE, etc. \

3. Endpoint Naming: App routes provide a way to assign names to specific URLs, known as endpoints. These endpoint names can be used in templates or other parts of the application to generate URLs dynamically.

4. Dynamic Routing: Flask allows for dynamic routing by using route parameters. Parameters can be specified in the URL pattern, and their values can be accessed within the associated function. This allows for more flexible and dynamic handling of different URL variations.

5. URL Generation: Flask's app routes enable URL generation within the application. By using the endpoint names and associated functions, Flask can generate URLs dynamically, making it easier to maintain consistent and correct URLs throughout the application.

6. Clean Code Structure: App routes help in organizing the application's code structure by separating different routes into distinct functions or view handlers. This enhances code readability, modularity, and maintainability.

7. RESTful APIs: App routes are commonly used to define routes for building RESTful APIs. Each route corresponds to a specific resource or endpoint, and the associated functions handle the corresponding CRUD operations.

```python
"""
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
Ans : 

url_for() function is used for URL building. It generates a URL for a specified endpoint by taking into account the associated function and any route parameters.
"""

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the Home Page'

@app.route('/about')
def about():
    return 'This is the About Page'

@app.route('/contact')
def contact():
    return 'Contact us at: contact@example.com'

@app.route('/dashboard')
def dashboard():
    home_url = url_for('home')
    about_url = url_for('about')
    contact_url = url_for('contact')

    # Displaying the generated URLs
    return f'''
    <p>Home URL: {home_url}</p>
    <p>About URL: {about_url}</p>
    <p>Contact URL: {contact_url}</p>
    '''

if __name__ == '__main__':
    app.run(host='0.0.0.0')
```
