```python
"""
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Ans :
1. Scatter Plot:
Scatter plots in Seaborn are created using sns.scatterplot().
They show the relationship between two continuous variables as individual data points, making it easy to identify patterns and correlations in the data.

2. Bar Plot:
To create bar plots in Seaborn, you can use sns.barplot().
Bar plots are used to represent the distribution of categorical data, showing the average or aggregated value of each category as a bar, helping to compare different categories easily.

3. Displot:
Displot is a figure-level interface for creating distribution plots, like histograms and kernel density plots.
It is created using sns.displot() and is useful for visualizing the distribution and characteristics of a single numeric variable.

4. Relplot:
Seaborn's sns.relplot() is used for creating relationship plots, allowing you to visualize the relationship between two numeric variables, often with additional categorical grouping.
It provides a flexible way to create scatter plots, line plots, or other plot types, facilitating quick exploration of data relationships.

5. Pairplot:
sns.pairplot() is used to create a grid of scatter plots or other types of plots for multiple numeric variables in the dataset.
It helps to visualize the pairwise relationships between variables, making it useful for understanding correlations and dependencies in a dataset with multiple numeric features.

"""
```

```python
"""
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
"""
import seaborn as sns
ds = sns.load_dataset('fmri')
sns.lineplot(x='timepoint', y='signal', data = ds, hue='event', style = 'region')
```

```python
"""
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
"""
import seaborn as sns
import matplotlib.pyplot as plt
ds = sns.load_dataset('titanic')
fig = plt.figure()
ax = fig.add_subplot(1,2,1)
sns.boxplot(x=ds['pclass'], y= ds['age'])
ax = fig.add_subplot(1,2,2)
sns.boxplot(x=ds['pclass'], y= ds['fare'])
fig.show()
```

```python
"""
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
"""
ds = sns.load_dataset('diamonds')
ds.head()
sns.histplot(x=ds.price, data= ds , hue = 'cut')
```

```python
"""
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
"""
ds = sns.load_dataset('iris')
sns.pairplot(data= ds, hue = 'species')

```

```python
"""
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
"""
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

# Create a pivot table of the flights data by year and month
pivot_table = flights.pivot_table(values="passengers", index="month", columns="year")
print(pivot_table)

# Plot the heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(pivot_table, cmap="Blues")
plt.title("Number of Flights by Month and Year")
plt.show()
```
