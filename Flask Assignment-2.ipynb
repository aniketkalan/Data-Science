```python
"""
Q1. Explain GET and POST methods.
Ans : GET and POST are two commonly used HTTP methods for handling requests.
In GET method, we can get the paraeters from request. These parameters are visible in request url
    Example : https://www.google.com/search?q=data+science

In POST method also we can get the parameters from the request but these parameters are not visible in the request url
POST method is more secured.
It is usually used in form submission

"""
from flask import Flask, request

app = Flask(__name__)

@app.route('/get_data', methods=['GET'])
def get_data():
    name = request.args.get('name')  
    return f"Return data for : {name}!"

if __name__ == '__main__':
    app.run(host = '0.0.0.0')
```

```python
# POST example:

from flask import Flask, request

app = Flask(__name__)

@app.route('/login', methods=['POST'])
def login():
    username = request.form['username']  # Accessing form data
    password = request.form['password']
    
    # Perform authentication logic
    if username == 'admin' and password == 'secret':
        return "Login successful!"
    else:
        return "Invalid credentials"

if __name__ == '__main__':
    app.run()
    
```

```python
"""
Q3. Why is redirect() used in Flask?
Ans : redirect() function is used to redirect the client's web browser to a different URL. 
It is often used after processing a request to redirect the user to a different page or route within the application.
"""
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/')
def index():
    # Redirect to the 'hello' route
    return redirect(url_for('redirected'))

@app.route('/redirected')
def redirected():
    return 'You are redirected here'

if __name__ == '__main__':
    app.run(host = '0.0.0.0')

```

```python
"""
Q4. What are templates in Flask? Why is the render_template() function used?
Ans :
In Flask whenever we want to render a html file then we use render_template function.
To use this function we need to import it from flask module
This allows us to separate the presentation logic from the application logic. 
Templates make it easier to manage and generate dynamic HTML pages by combining static HTML content with dynamic data.
"""

from flask import Flask, render_template


app = Flask(__name__)

@app.route("/")
def hello_world():
    return render_template('index.html' , name='Aniket', surname= 'Kalan')

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
