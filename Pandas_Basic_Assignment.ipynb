```python
"""
Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.
Ans :
A DataFrame in pandas is a two-dimensional tabular data structure with labeled axes.
DataFrames are commonly used to store and manipulate tabular data, such as data from spreadsheets, databases, and surveys.

A Series in pandas is a one-dimensional labeled array. It is a single column of a DataFrame.
Series are commonly used to store and manipulate tabular data that has a single dimension.

The main differences between DataFrames and Series are:

1. Dimensions: A DataFrame is two-dimensional, while a Series is one-dimensional.
2. Labels: A DataFrame has both row and column labels, while a Series only has row labels.
3. Heterogeneity: A DataFrame can have columns with different data types, while a Series must have all of its data in the same data type.
4. Size mutability: The size of a DataFrame can be changed by adding or dropping rows or columns, while the size of a Series cannot be changed.
"""
import pandas as pd
# Create a DataFrame
df = pd.DataFrame({'Name': ['Bam', 'Sham', 'Ram'], 'Age': [30, 25, 40]})

# Create a Series
s = pd.Series([30, 25, 40], name='Age')
print('Dataframe : ')
print(df)
print('\nSeries : ')
print(s)

```

```python
"""
Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?
Ans :
1. replace: This function is used to replace a specific value in a DataFrame with another value.
    For example, you could use df.replace('John Doe', 'Jane Doe') to replace all instances of John Doe with Jane Doe in a DataFrame.
2. drop: This function is used to drop a row or column from a DataFrame.
    For example, you could use df.drop('Name', axis=1) to drop the Name column from a DataFrame.
3. rename: This function is used to rename a row or column in a DataFrame.
    For example, you could use df.rename(columns={'Name': 'First Name'}, inplace=True) to rename the Name column to First Name in a DataFrame.
4. append: This function is used to append a row or column to a DataFrame.
    For example, you could use df.append({'Name': 'John Smith', 'Age': 35}, ignore_index=True) to append a new row to a DataFrame.
5. merge: This function is used to merge two DataFrames together.
    For example, you could use pd.merge(df1, df2, on='Name') to merge two DataFrames together based on the Name column.
"""
print(df)
df1 = df.replace('Alice', 'ALICE')
print(df1)

df2 = df.drop('Name', axis = 1)
print(df2)

df3 = df.rename(columns={'Name': 'First Name'})
print(df3)

df4 = df.append({'Name': 'John Smith', 'Age': 35},ignore_index=True)
print(df4)

df5 = pd.merge(df, df4, on='Name')
print(df5)
```
