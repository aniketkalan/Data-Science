```python
"""
Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.
Ans : Using the Exception class as the base for the custom exceptions in Python provides inheritance of useful features,
      aligns with standard practices, facilitates exception handling, ensures compatibility with existing constructs,
      and enhances code readability and maintainability.

"""
class MyException(Exception):
  pass
try:
  raise MyException("My custom exception")
except MyException as e:
  print(e)
```

```python
"""
Q2. Write a python program to print Python Exception Hierarchy.
"""

def print_exception_hierarchy(exception_class, indent=0):
    print(" " * indent + str(exception_class.__name__))
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

# Print the exception hierarchy starting from the base Exception class
print_exception_hierarchy(Exception)
```

```python
"""
Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
Ans : The ArithmeticError class in Python is a base class for exceptions that occur during arithmetic operations.
      Below are the errors defined in the ArithmeticError class:
      1. ZeroDivisionError
      2. ValueError / OverflowError
      3. FloatingPointError
      4. DecimalException

"""
# ZeroDivisionError
try:
  a = 10
  b = 10 / 0
except ZeroDivisionError as e:
  print("ZeroDivisionError raised - %s" %e)

# ValueError
import math
try:
  result = math.sqrt(-1)
  print(result)
except ValueError as e:
  print("ValueError raised - %s" %e)
```

```python
"""
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
Ans : The LookupError class in Python is used as a base class for all exceptions that are raised when an index or a key is not found for dictionary.
      It serves as a superclass for exceptions that occur when an index or key is not found during lookup operations on sequences or mappings.
      Two common exceptions derived from the LookupError class are KeyError and IndexError.

"""
# KeyError:
try:
  dict = {"Key1" : "val1" , "Key2" : "val2"}
  print(dict["key300"])
except KeyError as e:
  print("Key errorr occured %s " %e)

# IndexError:
try:
  lst = [1,2,3,4,5,6]
  print(lst[300])
except IndexError as e:
  print("IndexError occured %s " %e)
```
