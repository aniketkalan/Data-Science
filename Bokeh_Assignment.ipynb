```python
"""
Q1. How can you create a Bokeh plot using Python code?
"""
# Bokeh plot can be created as below
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers
output_file('test.html')
p = figure(title = 'test')
p.xaxis.axis_label = 'x'
p.yaxis.axis_label= 'y'
p.circle(flowers.sepal_length, flowers.sepal_width)
show(p)
```

```python
"""
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
Ans :
In Bokeh, glyphs are the geometric shapes used to represent data points or patterns on a plot.
They are the visual elements that display the actual data on the figure canvas.
Bokeh supports a wide range of glyphs, such as circles, squares, lines, bars, patches, and more.

"""
from bokeh.plotting import figure, show

x_values = [1, 2, 3, 4, 5]
y_values = [1, 3, 5, 7, 9]

p = figure(title="My Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add circles glyph
p.circle(x_values, y_values, size=10, color="blue", fill_color="white")

show(p)
```

```python
"""
Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

"""
from bokeh.plotting import figure, show

x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 1, 5, 3]

p = figure(title="Customized Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

p.circle(x_values, y_values, size=10, color="blue", fill_color="white", legend_label="Circles")

# Customize the title font size and alignment
p.title.align = "center"
p.title.text_font_size = "20pt"

# Customize axis tick labels
p.xaxis.major_label_orientation = "vertical"

# Customize the legend position
p.legend.location = "top_left"
p.legend.title = "Legend Title"
p.legend.label_text_font_size = "12pt"

# Remove grid lines
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

show(p)
```

```python
"""
Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?
Ans : A Bokeh server is a Python process that allows you to create interactive and dynamic data visualizations that can be updated in real-time.
It enables the creation of web applications with interactive Bokeh plots, where users can interact with the plots, and the data or visualizations can be updated based on user inputs or other external events.
The Bokeh server is a powerful tool for building interactive dashboards, data exploration tools, and real-time data monitoring applications.
"""
import bokeh.plotting as bp
from bokeh.server import Server

p = bp.figure()
p.line([1, 2, 3], [4, 5, 6])

server = Server(p, port=8000, num_workers=4)
server.start()
```

```python
"""
Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?
"""

import bokeh.plotting as bp
from bokeh.embed import components

# Create a Bokeh plot
p = bp.figure()
p.line([1, 2, 3], [4, 5, 6])

# Get the HTML code for the plot
script, div = components(p)

# Add the HTML code to the web page
app = Flask(__name__)

@app.route("/")
def index():
    return div

if __name__ == "__main__":
    app.run(debug=True)
```
